```python
aggregated_reviews = customer_reviews.groupby('product_id').agg(
    product_summaries=('summary_text', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    
    all_aspect_names=('aspect_name', lambda x: list(x.unique())), # Get a list of unique aspect names
    review_count=('doc_id', 'count'), 
    overall_sentiment= ('overall_sentiment', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    mentions_total=  ('mentions_total',lambda x: ' | '.join(x.dropna().astype(str).unique())),
    mentions_positive= ('mentions_positive',lambda x: ' | '.join(x.dropna().astype(str).unique())),
    negative= ('mentions_negative', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    doc_type= ('doc_type', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    snippet_1=('snippet_1_text', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    snippet_1_highlight= ('snippet_1_highlight', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    snippet_2=('snippet_2_text', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    snippet_2_highlight= ('snippet_2_highlight', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    snippet_3=('snippet_3_text', lambda x: ' | '.join(x.dropna().astype(str).unique())),
    snippet_3_highlight= ('snippet_3_highlight', lambda x: ' | '.join(x.dropna().astype(str).unique())),
).reset_index() 

print("\n--- Aggregated reviews data per product_id (from customers_review.csv) ---")
print(aggregated_reviews.head())
print(f"\nShape of aggregated_reviews: {aggregated_reviews.shape}")
```

```python
merged_df = pd.merge(documents, aggregated_reviews, on='product_id', how='left')

print("\n--- Merged DataFrame (all_documents.csv + aggregated reviews) ---")
# To display wide columns better
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 100) # Limit col width for readability here
print(merged_df)
print(f"\nShape of merged_df: {merged_df.shape}")
```

```python
output_filename = 'all_documents_merged.csv'
merged_df.to_csv(output_filename, index=False)
print(f"\nSuccessfully merged data saved to '{output_filename}'")
```

```python
import pandas as pd

# --- Configuration ---
REVIEWS_FILE = 'customers_review.csv'
ALL_DOCS_FILE = 'all_documents.csv'
OUTPUT_FILE = 'all_documents_with_expanded_concatenated_reviews.csv' # Updated output file name

GROUP_BY_COLUMN = 'product_id'
CONCAT_SEPARATOR = ' | ' # Separator for concatenated strings

# --- Load Data ---
try:
    reviews_df = pd.read_csv(REVIEWS_FILE)
    all_docs_df = pd.read_csv(ALL_DOCS_FILE)
except FileNotFoundError as e:
    print(f"Error loading files: {e}")
    print(f"Make sure '{REVIEWS_FILE}' and '{ALL_DOCS_FILE}' are in the script's directory or provide full paths.")
    exit()

print(f"--- Original '{REVIEWS_FILE}' (head) ---")
print(reviews_df.head(2))
print(f"Shape: {reviews_df.shape}\n")

print(f"--- Original '{ALL_DOCS_FILE}' (head) ---")
print(all_docs_df.head())
print(f"Shape: {all_docs_df.shape}\n")

# --- Step 1: Identify product_ids present in BOTH dataframes ---
if GROUP_BY_COLUMN not in reviews_df.columns:
    print(f"Error: Group by column '{GROUP_BY_COLUMN}' not found in '{REVIEWS_FILE}'.")
    exit()
if GROUP_BY_COLUMN not in all_docs_df.columns:
    print(f"Error: Group by column '{GROUP_BY_COLUMN}' not found in '{ALL_DOCS_FILE}'.")
    exit()

product_ids_in_reviews = set(reviews_df[GROUP_BY_COLUMN].unique())
product_ids_in_all_docs = set(all_docs_df[GROUP_BY_COLUMN].unique())
common_product_ids = list(product_ids_in_reviews.intersection(product_ids_in_all_docs))

if not common_product_ids:
    print("No common product_ids found between the two files. No concatenation will be performed.")
    all_docs_df.to_csv(OUTPUT_FILE, index=False)
    print(f"Saved '{ALL_DOCS_FILE}' to '{OUTPUT_FILE}' as no common products were found.")
    exit()

print(f"Found {len(common_product_ids)} common product_ids. Example(s): {common_product_ids[:min(5, len(common_product_ids))]}\n")

# --- Step 2: Filter reviews_df to only include reviews for these common product_ids ---
filtered_reviews_df = reviews_df[reviews_df[GROUP_BY_COLUMN].isin(common_product_ids)].copy()

if filtered_reviews_df.empty:
    print(f"After filtering, no review data remains for the common product_ids. Merging '{ALL_DOCS_FILE}' without new review data.")
    all_docs_df.to_csv(OUTPUT_FILE, index=False)
    print(f"Saved '{ALL_DOCS_FILE}' to '{OUTPUT_FILE}'.")
    exit()

print(f"--- Filtered '{REVIEWS_FILE}' (for common products only) (head) ---")
print(filtered_reviews_df.head(2))
print(f"Shape: {filtered_reviews_df.shape}\n")

# --- Step 3: Define aggregation logic (incorporating user's snippet and adding others) ---
CONCAT_LAMBDA = lambda x, sep=CONCAT_SEPARATOR: sep.join(x.dropna().astype(str).unique())
LIST_LAMBDA = lambda x: list(x.dropna().astype(str).unique())

# Check for 'doc_id' for counting, as specified in the user's snippet.
if 'doc_id' not in filtered_reviews_df.columns:
    print(f"Error: 'doc_id' column (for review_count) not found in '{REVIEWS_FILE}'. Exiting.")
    exit()

# Define all aggregations explicitly
aggregation_functions = {
    # Based on user's snippet (using robust lambdas and explicit source columns)
    'concatenated_summaries': ('summary_text', CONCAT_LAMBDA),
    'concatenated_snippet_1': ('snippet_1_text', CONCAT_LAMBDA), # Assuming snippet_1 means snippet_1_text
    'concatenated_snippet_2': ('snippet_2_text', CONCAT_LAMBDA), # Assuming snippet_2 means snippet_2_text
    'concatenated_snippet_3': ('snippet_3_text', CONCAT_LAMBDA), # Assuming snippet_3 means snippet_3_text
    'all_aspect_names': ('aspect_name', LIST_LAMBDA), # As a list of unique strings
    'review_count': ('doc_id', 'count'),

    # Adding "other things" also as concatenated strings
    'concatenated_overall_sentiment': ('overall_sentiment', CONCAT_LAMBDA),
    'concatenated_mentions_total': ('mentions_total', CONCAT_LAMBDA),
    'concatenated_mentions_positive': ('mentions_positive', CONCAT_LAMBDA),
    'concatenated_mentions_negative': ('mentions_negative', CONCAT_LAMBDA),
    'concatenated_doc_type': ('doc_type', CONCAT_LAMBDA),
    'concatenated_snippet_1_highlight': ('snippet_1_highlight', CONCAT_LAMBDA),
    'concatenated_snippet_2_highlight': ('snippet_2_highlight', CONCAT_LAMBDA),
    'concatenated_snippet_3_highlight': ('snippet_3_highlight', CONCAT_LAMBDA),
}

# --- Step 4: Validate that all source columns for aggregation exist ---
# Extract source column names from the aggregation_functions dictionary
source_columns_needed = list(set([val[0] for val in aggregation_functions.values()])) # Unique source columns

missing_source_cols = [col for col in source_columns_needed if col not in filtered_reviews_df.columns]

if missing_source_cols:
    print(f"Warning: The following source columns required for aggregation are missing from the filtered review data: {missing_source_cols}")
    print("Attempting to proceed by removing aggregations for missing columns.")
    # Filter out aggregations whose source columns are missing
    aggregation_functions = {
        key: (source_col, agg_func)
        for key, (source_col, agg_func) in aggregation_functions.items()
        if source_col in filtered_reviews_df.columns
    }
    if not aggregation_functions:
        print("Error: No valid aggregation functions remaining after checking for missing columns. Exiting.")
        exit()
    print("Updated aggregation functions based on available columns.")


# --- Step 5: Perform Aggregation on the filtered_reviews_df ---
print("\nPerforming aggregation...")
aggregated_reviews_df = filtered_reviews_df.groupby(GROUP_BY_COLUMN).agg(
    **aggregation_functions
).reset_index()

print(f"--- Aggregated review data (from common products only) (head) ---")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
pd.set_option('display.max_colwidth', 30) # Adjusted for potentially many columns
print(aggregated_reviews_df.head())
print(f"Shape: {aggregated_reviews_df.shape}\n")

# --- Step 6: Merge aggregated data with all_documents.csv ---
merged_df = pd.merge(all_docs_df, aggregated_reviews_df, on=GROUP_BY_COLUMN, how='left')

print(f"--- Merged DataFrame ('{ALL_DOCS_FILE}' + aggregated reviews for common products) (head) ---")
print(merged_df.head())
print(f"Shape: {merged_df.shape}\n")

# --- Save the result ---
try:
    merged_df.to_csv(OUTPUT_FILE, index=False)
    print(f"Successfully saved concatenated and merged data to '{OUTPUT_FILE}'")
except Exception as e:
    print(f"Error saving output file: {e}")
```
